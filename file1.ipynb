```python
import pandas as pd

# Load the already sampled dataset
data_sampled = pd.read_csv("datatransactions_3000.csv")

print("Data shape:", data_sampled.shape)
data_sampled.head()
```

```python
# Drop leakage column
data_sampled = data_sampled.drop(columns=["isFlaggedFraud"])

# Define features and target
X = data_sampled.drop(columns=["isFraud"])
y = data_sampled["isFraud"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
# Numeric and categorical features
num_features = ["amount", "oldbalanceOrg", "newbalanceOrig", "oldbalanceDest", "newbalanceDest"]
cat_features = ["type"]

# Column transformer
preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_features),
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_features)
    ]
)
```

```python
# Technique 1: Algorithmic (class_weight)
log_reg = LogisticRegression(max_iter=500, class_weight="balanced")

# Technique 2: Resampling (SMOTE)
smote = SMOTE(random_state=42)
log_reg_smote = ImbPipeline(steps=[("preprocessor", preprocessor),
                                   ("smote", smote),
                                   ("model", LogisticRegression(max_iter=500))])
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=500, class_weight="balanced"),
    "Random Forest": RandomForestClassifier(class_weight="balanced", random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric="logloss", scale_pos_weight=10)
}

results = {}

for name, model in models.items():
    pipe = Pipeline(steps=[("preprocessor", preprocessor), ("model", model)])
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    y_proba = pipe.predict_proba(X_test)[:,1]

    ap = average_precision_score(y_test, y_proba)
    results[name] = ap
    print(f"\n{name} - Average Precision: {ap:.4f}")
    print(classification_report(y_test, y_pred))
```

```python
best_model_name = max(results, key=results.get)
print("Best model:", best_model_name)

# Final model pipeline
if best_model_name == "Logistic Regression":
    final_model = Pipeline(steps=[("preprocessor", preprocessor),
                                  ("model", LogisticRegression(max_iter=500, class_weight="balanced"))])
elif best_model_name == "Random Forest":
    final_model = Pipeline(steps=[("preprocessor", preprocessor),
                                  ("model", RandomForestClassifier(class_weight="balanced", random_state=42))])
else:
    final_model = Pipeline(steps=[("preprocessor", preprocessor),
                                  ("model", XGBClassifier(use_label_encoder=False, eval_metric="logloss", scale_pos_weight=10))])

final_model.fit(X_train, y_train)
```

```python
# ===============================
# 1️⃣ Import Libraries
# ===============================
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, average_precision_score
from sklearn.ensemble import RandomForestClassifier
import shap
import joblib
import matplotlib.pyplot as plt
import os

# ===============================
# 2️⃣ Load Dataset
# ===============================
data_sampled = pd.read_csv("datatransactions_3000.csv")
print("Data shape:", data_sampled.shape)
data_sampled.head()

# ===============================
# 3️⃣ Prepare Features & Target
# ===============================
# Drop leakage column
data_sampled = data_sampled.drop(columns=["isFlaggedFraud"])

# Features & target
X = data_sampled.drop(columns=["isFraud"])
y = data_sampled["isFraud"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# ===============================
# 4️⃣ Preprocessing
# ===============================
num_features = ["amount", "oldbalanceOrg", "newbalanceOrig", "oldbalanceDest", "newbalanceDest"]
cat_features = ["type"]

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_features),
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_features)
    ]
)

# ===============================
# 5️⃣ Train Random Forest Pipeline
# ===============================
final_model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", RandomForestClassifier(class_weight="balanced", random_state=42))
])

final_model.fit(X_train, y_train)

# ===============================
# 6️⃣ Evaluate Model
# ===============================
y_pred = final_model.predict(X_test)
y_proba = final_model.predict_proba(X_test)[:,1]

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
print("Average Precision:", average_precision_score(y_test, y_proba))

# ===============================
# 7️⃣ SHAP Explanation
# ===============================
# Preprocess training data
X_train_preprocessed = final_model.named_steps["preprocessor"].transform(X_train)

# SHAP explainer (Random Forest + preprocessed data)
explainer = shap.TreeExplainer(final_model.named_steps["model"], X_train_preprocessed)
shap_values = explainer(X_train_preprocessed)

# Feature names
num_features = final_model.named_steps["preprocessor"].transformers_[0][2]
cat_features = final_model.named_steps["preprocessor"].transformers_[1][1].get_feature_names_out(["type"])
feature_names = np.concatenate([num_features, cat_features])

# Extract SHAP values for fraud class (class 1)
shap_values_fraud = shap_values.values[:, :, 1]

# Summary plot
shap.summary_plot(shap_values_fraud, X_train_preprocessed, feature_names=feature_names)

# Dependence plots for important features
shap.dependence_plot("amount", shap_values_fraud, X_train_preprocessed, feature_names=feature_names)
shap.dependence_plot("oldbalanceOrg", shap_values_fraud, X_train_preprocessed, feature_names=feature_names)
shap.dependence_plot("newbalanceOrig", shap_values_fraud, X_train_preprocessed, feature_names=feature_names)

# ===============================
# 8️⃣ Save Model
# ===============================
os.makedirs("models", exist_ok=True)
joblib.dump(final_model, "models/fraud_model.pkl")
print("Model saved as 'models/fraud_model.pkl'")
```
